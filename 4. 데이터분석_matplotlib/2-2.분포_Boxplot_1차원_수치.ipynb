## Boxplot

* 5가지 정보를 확인 할 수 있습니다. 
* 최솟값
* 제 1사분위수 - 25%
* 제 2사분위수(), 즉 중앙값
* 제 3 사분위수 -75%
* 최댓값
* 이상치(Outlier) 파악 가능합니다. 

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. 기본 스타일
## 폰트 설정 및 마이너스 부호 설정
#plt.rc('font', family='NanumSquare') #나눔 글꼴로 설정
plt.rc('font', family='Malgun Gothic') #맑은 고딕으로 설정
#plt.rc('font', family='AppleGothic') #맥
plt.rcParams['axes.unicode_minus'] = False #부호

plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 8

df = pd.read_csv('./data/house-price.csv')
df.head()
```

```python
#기술 통계 정보 확인 (수치형 데이터)
df.describe()
```

```python
y = df['분양가'].values
plt.boxplot(y) 
plt.show()
```

```python
y = df['월'].values
plt.boxplot(y) 
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. 기본 스타일
## 폰트 설정 및 마이너스 부호 설정
#plt.rc('font', family='NanumSquare') #나눔 글꼴로 설정
plt.rc('font', family='Malgun Gothic') #맑은 고딕으로 설정
#plt.rc('font', family='AppleGothic') #맥
plt.rcParams['axes.unicode_minus'] = False #부호

plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 8


# Load the Iris dataset from seaborn
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
iris.describe()
```

```python
# 붓꽃 데이터의 각 개별 피처의 분포를 분석하시오.
# 이상치가 존재하는 피처는?

y = iris['sepal_length'].values
plt.boxplot(y) 
plt.show()
```

```python
y = iris['sepal_width'].values
plt.boxplot(y) 
plt.show()
```

```python
y = iris['petal_length'].values
plt.boxplot(y) 
plt.show()
```

```python
y = iris[['sepal_length','sepal_width','petal_length' ,'petal_width']]
plt.boxplot(y, labels = ['sepal_length','sepal_width','petal_length' ,'petal_width'])
plt.boxplot(y) 
plt.title('iris 데이터 분포 분석')
plt.show()
```
