````python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 기본 스타일
## 폰트 설정 및 마이너스 부호 설정
#plt.rc('font', family='NanumSquare') #나눔 글꼴로 설정
plt.rc('font', family='Malgun Gothic') #맑은 고딕으로 설정
#plt.rc('font', family='AppleGothic') #맥
plt.rcParams['axes.unicode_minus'] = False #부호

plt.rcParams['figure.figsize'] = (4, 3)
plt.rcParams['font.size'] = 8
````

````python
# 연비데이터셋
mpg = sns.load_dataset("mpg")
````

````python
mpg.info()
````

```
mpg: 자동차 모델의 갤런당 마일(연비).
실린더: 자동차 엔진의 실린더 수.
배기량: 세제곱인치로 측정한 자동차 엔진의 배기량.
마력: 자동차 엔진의 마력.
무게: 차량의 무게(파운드).
가속: 0에서 60mph까지의 가속 시간(초).
model_year: 자동차 모델이 제조된 연도(예: 70은 1970년에 해당).
원산지: 범주형 변수로 표시되는 자동차 제조업체의 원산지(1: 미국, 2: 유럽, 3: 일본).
car_name: 자동차 모델의 이름입니다.
```

````python
mpg.head()
````

````python
mpg['weight']
````

````python
#무게와 연비와의 관계는
x=mpg['weight']
y=mpg['mpg']
plt.scatter(x=x, y=y, color = 'Green')
plt.show()
# 강한 음의 상관관계이기 때문에 연비 개선을 위해서는 무게를 잡을 필요가 있다.
````

````python
#연비와 배기량
x=mpg['displacement']
y=mpg['mpg']
plt.scatter(x=x, y=y, color = 'red')
plt.show()
````

````python
#실린더 개수와 연비와의 관계
x=mpg['cylinders']
y=mpg['mpg']
plt.scatter(x=x, y=y, color = 'black')
plt.show()
````

````python
sns.heatmap(mpg.corr(numeric_only=True),annot=True, annot_kws={'size':7},\
                    cmap=plt.cm.RdYlBu_r, fmt=".2f", linewidth=.5,vmin=-1.0,square=False)
````
