## 판다스 그룹핑

1) mpg: 연비
2) cylinders: 실린더 개수
3) displacement: 배기량
4) horsepower: 마력
5) weight: 무게
6) acceleration: 엔진이 초당 얻을 수 있는 가속력
7) model year: 출시 년도
8) origin: 제조 장소(1: 미국 USA, 2: 유럽 EU, 3: 일본 JPN)
9) car name: 자동차 이름

```python
import pandas as pd
import seaborn as sns
```

```python
#내장 데이터 셋 로드
df = sns.load_dataset("mpg")  #자동차의 연비와 관련된 데이터
df.head(20)
```

```python
df.info()
```

## 1. 년도 별 관측된 데이터 개수 보기

```python
df.groupby(by='model_year').count()
```

## 2. 제조국별 생산된 실린더 합계

```python
df.groupby(by='origin').sum()['cylinders']
```

```python
df.groupby(by='origin')['cylinders'].sum()
```

## 2. 연도에 따른 제조국별 실린더 합계보기

```python
# 이중 그룹핑
# 년도별 제조국 무
df.groupby(by = ['model_year', 'origin'])['cylinders'].sum()
```

## 3. 평균값으로 보기

```python
df.groupby(by = ['model_year', 'origin'])['cylinders'].mean()
```

## 4. 여러 통계 함수를 적용하여 보기

```python
#마력
df.groupby(by = ['model_year', 'origin'])['cylinders'].aggregate(['mean', 'median', 'max', 'min'])
```
