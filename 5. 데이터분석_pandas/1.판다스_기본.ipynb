### 판다스 라이브러리
- 테이블형 데이터를 다룰 수 있는 다양한 기능을 가진 라이브러리
  - 파이썬 데이터 분석을 위해 기본적으로 사용하는 라이브러리임
- raw data를 데이터 분석 전과정을 위해 사용할 수 있도록 변환하는 데이터 전처리에도 많이 사용됨
  - raw data: 아직 데이터 분석을 위해 정제되지 않은 기본 데이터를 의미함
    - 보통 데이터 분석 목적에 맞지 않은 불필요한 데이터가 있거나, 데이터가 없는 열들이 포함됨

## pandas 데이터 타입
- pandas 데이터 타입은 파이썬과 다름
  - dtype 으로 불리우며, 주요 데이터 타입은 다음과 같음
    - object 는 파이썬의 str 또는 혼용 데이터 타입 (문자열)
    - int64 는 파이썬의 int (정수)
    - float64 는 파이썬의 float (부동소숫점)
    - bool 는 파이썬의 bool (True 또는 False 값을 가지는 boolean)
    - 이외에 datetime64 (날짜/시간), timedelta[ns] (두 datatime64 간의 차) 도 활용됨

> 가끔 data type 때문에 에러가 나는 경우가 있으므로, 데이터 타입에 대한 이해 및 데이터 타입 변경 기능을 알아둬야 함

````python
#object 타입
sd = pd.Series(['dave', 'alex', 'amir'])
sd
````

````python
# 정수형 타입
sd = pd.Series([1,2,4])
sd
````

````python
# object 타입
sd = pd.Series(['안녕', 2, 4])
sd
````

### 데이터 하나에 접근하는 방법

* df.loc[행 라벨명,컬럼명]
* df.iloc[행 인덱스,컬럼 인덱스]

````python
df = pd.DataFrame({
    "년도": [2020, 2021, 2022,2023],    
    "미국": [2.1, 2.2, 2.3,2.4],
    "한국": [0.4, 0.5, 0.45,1.3],
    "중국": [10, 13, 15,9],
    "일본": [6, 9, 2,1]       
})
df = df.set_index('년도')
df
````

````python
df.index
````

````python
#미국의 2.1값 가져오기
df.loc[2020, '미국']

````

````python
df.iloc[0,0]
````

#### 데이터 하나 수정

````python
#2020년, 미국 2.8로 수정
df.loc[2020, '미국'] = 2.8
df
````

````python
#2020년, 미국 2.1로 수정 (iloc)
df.iloc[0,0] = 2.1
df
````

#### 데이터 여러개에 접근하는 방법

````python
#2020년의 미국,중국 데이터 추출
df.loc[2020, ['미국', '중국']]
````

````python
#2020년의 미국,중국 데이터 추출 (프레임으로)
df.loc[[2020], ['미국', '중국']]
````

````python
#2021~2023 미국, 중국의 점수를 표시
df.loc[2021:2023, ['미국', '중국']]

````

````python
#2020년~2023년 미국, 중국의 점수를 표시
df.loc[:, ['미국', '중국']]
````

````python
# 2022년도 데이터 추출
df.loc[[2022], :]
````

````python
# 2021,2022년도 데이터 추출(인덱스 기준)
df.loc[2021:2022, :]
````

### 데이터 선택-조건활용

````python
df = pd.DataFrame({
    "년도": ['2020','2021', '2022', '2023'],    
    "미국": [2.1, 2.2, 2.3,2.4],
    "한국": [0.5, 0.4, 0.45,0.21],
    "중국": [17, 13, 15, 14],
    "등급": ['양호','위험','양호','양호']
})
df
````

````python
#비교조건으로 데이터 선택-> 시리즈타입
cond = df['년도'] > '2021'
cond
````

````python
#마스킹 연산으로 데이터프레임으로 표현
# 값이 Ture인 값들만 추출됨.
df[cond]
````

````python
df[df['년도'] > '2021']
````

````python
#원하는 문자열이 있는 데이터 선택
cond = df['등급']=='양호'
df[cond]
````

````python
# and 연산
cond = (df['등급']=='양호') & (df['한국'] > 0.4)
df[cond]
````

````python
# OR 연산
cond = (df['등급']=='양호') | (df['한국'] > 0.4)
df[cond]
````

````python
cond = df['등급']=='양호'

df.loc[cond,:  ]
````

````python
cond = df['등급']=='양호'
df.loc[cond, ['미국', '중국']]
````

````python
cond = df['등급']=='양호'
df[cond][['미국', '중국']]
````
