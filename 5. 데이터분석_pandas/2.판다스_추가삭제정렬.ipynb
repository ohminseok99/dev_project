## 컬럼 추가

````python
import pandas as pd
````

````python
df = pd.DataFrame({
    "학번": ['A1021','A1022', 'A1023', 'A1024'],  
    "이름": ['김용명','다나카','유재석','이용진'], 
    "수학": [92, 48, 73, 89],
    "영어": [88, 75, 45, 98],
    "국어": [69, 96, 100,74],
})
df
````

````python
#새로운 컬럼 추가
df['출석여부'] = True
df
````

````python
#df.insert(컬럼인덱스, 컬럼명, 값)

df.insert(2,'수학경시출전',df['수학']>75)
df
````

````python
#컬럼간 연산 후 새 컬럼 만들기
df['총점'] = df['수학'] + df['영어'] + df['국어']
df 
````

## 컬럼 삭제

````python
#df.drop(컬럼이름,axis=1,inplace=True)
# df.drop('수학경시출전',axis=1) <- 이걸 실행했을 때는 원본이 바뀌는 게 아니라
# 그냥 그때만 일시적으로 바뀌는 것(프로비전)
# 그래서 df를 출력했을 때는 원본이 나옴.
df.drop('수학경시출전',axis=1)
df
````

````python
# 원본까지 바꾸려면 뒤에 inplac = True 넣어줘야 함.
df.drop('수학경시출전',axis=1,inplace=True)
df
````

````python
# 출석여부 삭제
df.drop('출석여부', axis = 1, inplace=True)
df
````

````python
import seaborn as sns
tit_df = sns.load_dataset("titanic")
````

````python
tit_df.info()
````

````python
#family(가족)의 총합은 sibsp컬럼과 parch의 합산
tit_df['familySize'] = tit_df['sibsp'] + tit_df['parch']
tit_df.head(5)
````

````python
#FamilySize 삭제
tit_df.drop('familySize', axis=1, inplace=True)
tit_df
````

````python
#parch 뒤에 FamilySize 삽입
tit_df.insert(6,'FamilySize',tit_df['sibsp'] + tit_df['parch'])
tit_df
````

````python
#fare가 100 초과일 경우 RICH 컬럼에 True로 삽입
tit_df.insert(7, 'RICH', tit_df['fare'] > 100)
tit_df.head()
````

````python
#fare, age기준으로 내림차순 정렬
tit_df.sort_values(by=['fare','age'],ascending=False)
````

````python
# 정렬 후 상위 5명 중에서 
# 여성이면서 혼자 탑승(alone = True)인 사람은 결국 타이타닉호에서 생존(survived == 1) 했나요?
a = tit_df.sort_values(by=['fare','age'],ascending=False)

cond1 = (a['sex'] == 'female') & (a['alone'] == True)
cond1
````
