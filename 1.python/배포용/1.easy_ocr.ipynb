```python
import easyocr
```

```python
reader = easyocr.Reader(['en','ko'], gpu=False) # this needs to run only once to load the model into memory
result = reader.readtext('./receipt2.jpg')
result
```

위가 어떤 구조로 되어 있나면
일단 복합리스트임
index 값이 0, 1, 2 세 개고, 타입은 0은 [] 1은 '', 2는 float임 
즉 우리가 필요한 부분은 문자열이고
그 말은 [1]이 필요한 것.


```python
for i in result:
    print(i[1])
```

```python
for idx, i in enumerate(result):
    print(idx, i[1])
```

```python
when = result[10][1]
where = result[3][1]
money = result[20][1]
```

```python
# 언제, 어디서, 얼마를 사용했는지 출력

print("날짜:", when)
print("장소:", where)
print("금액:", money)
```
