* 다양한 용어가 존재합니다.

### 기본값 정의

* 함수를 정의 할때 매개변수의 기본값까지 정의된 경우, 함수 호출 시 해당 매개변수를 생략하면 기본으로 지정된 값을 사용한다.
* 기본값 이외의 값을 적용하고 싶을 경우 함수 호출시 매개변수 값을 지정하면 된다.  

````python
# 주 근무 시간에 따른 주급(만원) 계산 함수 정의 
def week(hours):
    return hours * 10000
````

````python
week(10)
````

````python
week(23)
````

````python
# 매개변수에 기본값 정의
# 그냥 디폴트 값을 설정해준 것이고 사용자가 매개변수 값을 바꿀 수 있음.
def week(hours=40):
    return hours * 10000
````

````python
week()
````

````python
week(12)
````

### 필수 매개변수와 옵션 매개변수

* 필수는 반드시 제공해야하는 값
* 옵션은 기본 값이 제공된 경우

````python
def info(name, add, age=14, hobby='game'):
    print("이름:", name)
    print("주소:", add)
    print("나이:", age)
    print("취미:", hobby)
````

````python
info('조인성', '강남', 25, 'sing')
````

### 매개변수 개수를 모르는 경우

* 함수를 정의할 때 복수로 입력되는 매개변수 명 앞에 *를 붙여서 정의

````python
def birthday(*names):
    for name in names:
        print('{0}님을 생일파티에 초대합니다.'.format(name))
````

````python
birthday('천우희', '김채원', '비비', '김다미')
````

````python
def birthday(dt, *names):
    for name in names:
        print('{0}일 {1}님을 생일파티에 초대합니다.'.format(dt, name))
````

````python
birthday('2023/07/19', '천우희', '김채원', '비비', '김다미')
````

````python
birthday('2023/07/19')
````

## 연습문제

VIP 초대장 문구를 자동으로 작성하려고 합니다. 
아래와 같은 포맷에 맞춰 문구를 출력하세요.

```
안녕하세요 김수현 님
이번 저희 2023년 런칭쇼에 김수현 님을 초대합니다.
쿠폰No. 1
김수현님 감사합니다.
```

````python
vip_list=['김수현','유아인','차은우','박보검']

num = 0
for custom in vip_list:
    result = ''
    num += 1

    greet = f"안녕하세요 {custom}님\n"
        
    content = f"이번 저희 2023년 런칭쇼에 {custom}님을 초대합니다.\n"
    
    coupon = f"쿠폰No. {num}\n"
    last = coupon + f"{custom}님 감사합니다.\n"
    result = greet + content + last 

    print(result)
````

* 함수 만들어 보기

````python
# 함수를 기능 단위로 끊어서 만들기
# 그 기능 단위로 사용자 정의 함수를 만들어 보기
# 함수를 내가 호출해서 초대장 만들기 프로그램을 구현
````

````python
def invite(name, coupon, year = '2023'):
    print('안녕하세요. {0}님'.format(name))
    print('이번 저희 {0}년 런칭쇼에 {1}님을 초대합니다.'.format(year, name))
    print('쿠폰No. {0}'.format(coupon))
    print('{0}님 감사합니다.'.format(name))
    print()

    
````

````python
vip_list=['천우희', '김채원', '김다미', '비비']

for idx, vip in enumerate(vip_list):
    invite(vip, idx + 1)

````

## lambda

* lambda는 함수를 생성할 때 사용하는 예약어로 def와 동일한 역할을 한다. 
* 보통 함수를 한줄로 간결하게 만들 때 사용한다. 
* 함수 이름을 붙이지 않고 일시적으로 함수를 작성하고자 할 때 사용한다.
* 우리말로는 "람다"라고 읽고 def를 사용해야 할 정도로 복잡하지 않거나 def를 사용할 수 없는 곳에 주로 쓰인다.

```python

lambda 매개변수 : 표현식

````python
def plus2(num):
    return num + 2

plus2(2)
````

````python
lf = lambda num : num+2
lf(2)
````

## 변수 유효 범위

* 전역변수(global variable) : 프로그램 전체에서 접근이 가능
* 지역번수(local variable): 함수내에서 선언된 변수 -> 함수 내에서만 접근 가능

### 전역변수 VS 지역변수

````python
a="OUT"

def my_func():
    a = 'IN'
    print(a) #지역변수 a를 출력
````

````python
my_func()

print(a)
````

````python
def my_func2():
    print(a) #전역변수 값 출력

my_func2()
````

````python
a = "Out"
def my_func2():
    global a   #전역변수 값을  함수 내에서 수정하려면 global을 사용해야 함.
    a ='Update'  
````

````python
my_func2()

a
````
