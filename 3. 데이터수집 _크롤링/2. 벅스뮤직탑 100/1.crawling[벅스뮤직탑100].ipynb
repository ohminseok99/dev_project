## 벅스뮤직 탑100 크롤링 후 저장

```python
import requests
from bs4 import BeautifulSoup  
import pandas as pd


url = 'https://music.bugs.co.kr/chart'
```

```python
# 요청 requests
res = requests.get(url)
# 파싱
soup = BeautifulSoup(res.content,'html.parser')

#  순위
#CHARTrealtime > table > tbody > tr:nth-child(1) > td:nth-child(4) > div > strong
#CHARTrealtime > table > tbody > tr:nth-child(2) > td:nth-child(4) > div > strong
#score = '#CHARTrealtime > table > tbody > tr > td:nth-child(4) > div > strong'

# 곡
#CHARTrealtime > table > tbody > tr:nth-child(1) > th > p > a
#song = '#CHARTrealtime > table > tbody > tr > th > p > a'

# 가수
#CHARTrealtime > table > tbody > tr:nth-child(1) > td:nth-child(8) > p > a
#singer = '#CHARTrealtime > table > tbody > tr > td:nth-child(8) > p > a'

#CHARTrealtime > table > tbody > tr:nth-child(1)
#CHARTrealtime > table > tbody > tr:nth-child(1) > td:nth-child(4) > div > strong
#CHARTrealtime > table > tbody > tr:nth-child(1) > th > p > a
#CHARTrealtime > table > tbody > tr:nth-child(1) > td:nth-child(8) > p > a

# 이게 순위기 때문에 개별적으로 접근하기 보다는 1 ~ 100위까지 같은 구조로 되어 있어서
# 같은 부분까지만 css에 저장하고 나머지 뒤에 것들은 순위, 곡, 가수에 넣어주면 됨.

css = '#CHARTrealtime > table > tbody > tr'
data = soup.select(css)

rank_lst = []
song_lst = []
singer_lst = []

for i in data:
    rank = i.select_one('td > div > strong').text
    song = i.select_one('th > p > a').text
    singer = i.select_one('td > p > a').text

    print(rank, song, singer)

    rank_lst.append(rank)
    song_lst.append(song)
    singer_lst.append(singer)

#print(len(rank_lst), len(song_lst), len(singer_lst))


bugs_df = pd.DataFrame({'순위':rank_lst,'제목':song_lst,'가수':singer_lst})
bugs_df

bugs_df.to_csv('./data/벅스탑100크롤링_예제.csv', encoding = 'utf-8', index = False)
# csv 파일로 받아줌.
```
